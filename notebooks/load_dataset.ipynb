# Control the cache_dir

If you want to control "where" the dataset is cached locally, you can specify a cache dir

```python
import os

# Create a cache subdir based on the dataset path
CACHE_ROOT = "/cache"
remote_dir = "s3://visionlab-datasets/imagenet100/pre-processed/s256-l512-jpgbytes-q100-streaming/val"
cache_subdir = remote_dir.replace("s3://visionlab-datasets/", "").rstrip("/")
cache_dir = os.path.join(CACHE_ROOT, cache_subdir)
os.makedirs(cache_dir, exist_ok=True)
```

```python
ds = StreamingDataset(input_dir=Dir(path=cache_dir, url=remote_dir), decode_images=True)
ds
```

```python
sample = ds[0]
sample.keys()
```

```python
sample['image']
```
